```python
import requests  # Call library requests and bs4 
from bs4 import BeautifulSoup 
# requests : لتحميل صفحه الويب
# BeautifualSoup:لاستخراج البيانات من HTML
```

```python
response=requests.get("https://wuzzuf.net/search/jobs/?q=machine%20learning&a=navbg+") # معناه انك بعت ريكويست
response  # والريكوست نجح واستقبلت ريسبونس
```

```python
soup=BeautifulSoup(response.content,"lxml")  # to show content inside response باختصار عشان تجيب المحتوي بتاع الصفخه
soup   # lxml دا parser هوا المسؤل عن ال سيرش جوا الكونتت
```

```python
title=soup.find("h2",{'class':'css-m604qf'})
title   # دا كدا جاب البيانات بس موصلتش للي انا عايزه بالظبط
```

```python
title.text   # كدا انا وصلت للنتيجه اللي انا عايزها بالظبط
```

```python
titles=soup.find_all("h2",{'class':'css-m604qf'})
titles  # الكود دا هيرجع كل التاج اللي ف الصفحه 
```

```python
titles_lst=[title.text for title in titles]
titles_lst
```

```python
link_lst=[titles[0].a["href"] for link in titles]
link_lst 
```

```python
occupations = soup.find_all('div', {'class': 'css-1lh32fc'})  
# البحث عن جميع العناصر <div> التي تحتوي على الكلاس 'css-1lh32fc' وتخزينها في قائمة
occupations
```

```python
occupations=soup.find_all('div',{'class':'css-1lh32fc'})  # the same result of the last code but in their result inside result
occupations
```

```python
occupations[0] # access the first element in list
```

```python
occupations[0].text # access the first element & string in list
```

```python
occupations[0].text[:9] + '/' + occupations[0].text[9:]  
# access the first element in the list, split the text at index 9, and join with '/'
```

```python
occupations_lst=[occupation.text for occupation in occupations] # return all job describtion in list
occupations_lst
```

```python
companies=soup.find_all('a',{"class":"css-17s97q8"})  # return company name & and link for job
companies
```

```python
companies_lst = [company.text for company in companies] # to access companies name
companies_lst
```

```python
companies_lst=[company.text.replace('-','') for company in companies] # return companies name without column or spaces 
companies_lst
```

```python
details=soup.find_all('a',{'class':"css-o171kl"})
details[2].text
```

```python
details_lst=[detail.text for detail in details] # return all job title by describe
details_lst
```

```python
scraped_data={}
scraped_data['Title']=titles_lst
scraped_data['link']=link_lst 
scraped_data['Occupation']=occupations_lst
scraped_data['Company']=companies_lst
scraped_data['details']=details_lst
scraped_data
```

```python
import pandas as pd
df=pd.DataFrame(scraped_data)
df
```
